```python
import pandas as pd

data = pd.read_csv('data/2019.csv')
data
```

```python
orig_data = data[['Country or region', 'GDP per capita', 'Score']]
```

```python
data = data[['GDP per capita', 'Score']]
data
```

```python
x = data[data.columns[-2]]
y = data[data.columns[-1]]
x, y
```

```python
x_2d = x.to_numpy().reshape(x.shape[0], 1)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_2d, y)
```

```python
y_preds = lin_reg.predict(x_2d)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, s=2)
plt.plot(x, y_preds)
```

# Calculating Outliers from the line trend
Percentage threshold
1. Get all the distances of the instances from the line
2. Sort them in ascending order
3. Take the last %percentage% of them and mark them as anomalies

```python
data.values
x.to_numpy().reshape(-1,1)
```

```python
y_preds.reshape(-1,1)
```

```python
import numpy as np

preds = np.concatenate((x.to_numpy().reshape(-1,1), y_preds.reshape(-1,1)), axis=1)
```

```python
actual = data.to_numpy()
```

```python
# distance in the y axis only

# only gets the y distance
dists = (actual - preds).reshape(-1,1)[1::2]
```

```python
# get y distances in 1d array
dists1 = (actual - preds).reshape(1,-1)[0][1::2]
dists1 = abs(dists1)
```

```python
orig_data = orig_data.assign(pred=y_preds)
orig_data = orig_data.assign(distance=dists1)
```

```python
orig_data = orig_data.sort_values(by='distance', ascending=False)
```

```python
percentage = 0.1
```

```python
length = int(len(orig_data) * percentage)+1
orig_data = orig_data.iloc[:length]
```

```python
plt.scatter(x, y, s=2)
plt.plot(x, y_preds)
plt.scatter(orig_data[['GDP per capita']], orig_data[['Score']], marker='*', c='r', s=10)
```
